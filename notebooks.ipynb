## Code blocks and outputs

Jupyter Book will also embed your code blocks and output in your book.
For example, here's some sample Matplotlib code:

```python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
import numpy as np
plt.ion()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

N = 10
data = [np.logspace(0, 1, 100) + np.random.randn(100) + ii for ii in range(N)]
data = np.array(data).T
cmap = plt.cm.coolwarm
rcParams['axes.prop_cycle'] = cycler(color=cmap(np.linspace(0, 1, N)))


import seaborn as sns
import pandas as pd

# Create a DataFrame from the data array with column names
df = pd.DataFrame(data, columns=[f'Series_{i}' for i in range(N)])

# Create the line plot using seaborn
plt.figure(figsize=(10, 5))
sns.set_palette(sns.color_palette("coolwarm", n_colors=N))
for col in df.columns:
    _ = sns.lineplot(data=df[col], label=col)
plt.legend(loc='upper center', ncol=5)
```
